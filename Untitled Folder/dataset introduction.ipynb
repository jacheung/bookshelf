```python
# loading discharge constant power (CPD) data
discharge_class = PSeriesDischargeCPDDataset()
discharge = discharge_class.load_data(data_filepaths=files, para_filepaths=[])
df_discharge = discharge[0]
df_discharge
```

```python
# loading charge data
charge_class = PSeriesChargeCCCVCDataset()
charge = charge_class.load_data(data_filepaths=files, para_filepaths=[])
df_charge = charge[0]
df_charge
```

### 1.3 basic feature engineering

```python
i_fe = IncrementalCalculationFeatureGenerator()
i_df = i_fe.generate_features(df_discharge)
i_df = pd.DataFrame(i_df[0], columns=i_df[1]['column_headers'])
```

```python
i_df
```

```python
plt.figure(figsize=(8,8))
for x in i_df['File Name'].unique():
    mask = i_df['File Name'] == x
    plt.plot(i_df.loc[mask, 'Cyc#'], i_df.loc[mask, 'V_at_dQdV_peak'], alpha=0.5)
    
plt.xlabel('Cycle #')
plt.ylabel('V_at_dQdV_peak')
plt.tight_layout()
plt.show
```
