```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('./remotasks/candy-data.csv')
df.set_index('competitorname', inplace=True)
df['winpercent'] = df['winpercent']/100
df.dtypes
```

```python
sns.heatmap(df.sort_values(by='winpercent'))
```

```python
values = pd.DataFrame(df.corr()['winpercent'].sort_values())
values.drop(labels=['winpercent'], inplace=True)

sns.barplot(data=values, x=values.index, y='winpercent', color='gray')
plt.xticks(rotation=90)
plt.show()
```
