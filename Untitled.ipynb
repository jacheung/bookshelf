```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
house_cost = 700000
yearly_increase = 1.01
price = []

for year in np.arange(0,31):
    price.append(house_cost * (yearly_increase ** year))


price
```

```python
plt.figure(figsize = (8,8))
plt.rcParams.update({'font.size' : 16})
plt.plot(np.arange(0,31),price)
plt.ylabel('price ($)')
plt.xlabel('year')
plt.show()
```
