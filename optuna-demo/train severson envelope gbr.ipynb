#### b. hyperparameter tuning
build GBR model (choosing because we want prediction intervals; XGB can't provide that)  
use genetic algorithms to select hyperparameters


```python
# clean up empty values/nan values
lookup_df = pd.read_csv('inference_missing_lookup.csv').astype(float).set_index('Cyc#')
dirty_columns = test_X.columns[test_X.isna().sum() > 0]
for cols in dirty_columns:
    dirty_indices = [idx for idx, x in enumerate(test_X[cols].isna()) if x]
    for idx in dirty_indices:
        dirty_idx_cyc = test_X['Cyc#'].iloc[idx]
        fill_value = lookup_df.loc[dirty_idx_cyc, cols]
        test_X[cols].iloc[idx] = fill_value
```

```python
study = optuna.create_study(study_name='severson-envelope-gbr',
                            storage='postgresql://localhost/optunadb',
                            direction='minimize')

def objective(trial):
    param={
        'n_estimators': trial.suggest_int('n_estimators', 10, 500),
        'min_samples_leaf': trial.suggest_int('min_samples_leaf', 50, 200),
        'max_depth': trial.suggest_int('max_depth', 10, 100),
        'learning_rate': trial.suggest_float('learning_rate', 0.001, 0.1)
    }
    model = GradientBoostingRegressor(loss='quantile', random_state=123,
                                     validation_fraction=0.2, n_iter_no_change=5,**param)
    model.fit(train_X, train_Y.values.ravel())
    pred_y = model.predict(test_X)
    rmse = np.sqrt(np.mean((pred_y-test_Y.values)**2))
    return rmse

study = optuna.load_study(study_name='severson-envelope-gbr',
                          storage='postgresql://localhost/optunadb')
study.optimize(objective, n_trials=32)
optuna.visualization.plot_param_importances(study)
```

```python
import matplotlib.pyplot as plt
# set up test X and Y
test_X = X_df[X_df['File Name'].isin(test_files)]
test_X_mapping = test_X[['File Name', 'Cyc#']].reset_index(drop=True)
test_X = test_X.drop(columns='File Name').astype(float)
test_Y = Y_df[Y_df['File Name'].isin(test_files)]

# clean up empty values/nan values
lookup_df = pd.read_csv('inference_missing_lookup.csv').astype(float).set_index('Cyc#')
dirty_columns = test_X.columns[test_X.isna().sum() > 0]
for cols in dirty_columns:
    dirty_indices = [idx for idx, x in enumerate(test_X[cols].isna()) if x]
    for idx in dirty_indices:
        dirty_idx_cyc = test_X['Cyc#'].iloc[idx]
        fill_value = lookup_df.loc[dirty_idx_cyc, cols]
        test_X[cols].iloc[idx] = fill_value

# make predictions
yhat = gbr_best.predict(test_X)
yhat_low = gbr_low.predict(test_X)
yhat_high = gbr_high.predict(test_X)
pred_df = pd.concat([test_X_mapping, pd.DataFrame({'Retention(%)_yhat': yhat,
                                                   'Retention(%)_yhat_5': yhat_low,
                                                   'Retention(%)_yhat_95': yhat_high})], axis=1)
compare_df = pred_df.merge(test_Y, on=['File Name', 'Cyc#'])

mape = np.mean(abs(compare_df['Retention(%)_yhat'] - compare_df['Retention(%)']))

plt.figure(figsize=(12, 8))
mean_y = compare_df.groupby('Cyc#')[['Retention(%)']].agg('mean')
mean_yhat = compare_df.groupby('Cyc#')[['Retention(%)_yhat']].agg('mean')
mean_yhat_low = compare_df.groupby('Cyc#')[['Retention(%)_yhat_5']].agg('mean')
mean_yhat_high = compare_df.groupby('Cyc#')[['Retention(%)_yhat_95']].agg('mean')
plt.plot(mean_y.index, mean_y, color='k', linewidth=2, label='y_mean')
plt.plot(mean_yhat.index, mean_yhat, color='b', linewidth=2, label='yhat_mean')
plt.plot(mean_yhat.index, mean_yhat_low, 'b--', linewidth=1)
plt.plot(mean_yhat.index, mean_yhat_high, 'b--', linewidth=1)
for x in test_files:
    mask = compare_df['File Name'] == x
    plt.plot(compare_df.loc[mask, 'Cyc#'], compare_df.loc[mask, 'Retention(%)'], color='k', alpha=.2)
    plt.plot(compare_df.loc[mask, 'Cyc#'], compare_df.loc[mask, 'Retention(%)_yhat'], color='blue', alpha=.2)
plt.title(f'GBR model on P42 test; MAPE:{mape.round(2)}')
plt.xlabel('Cyc#')
plt.ylabel('Retention(%)')
plt.legend()
plt.show()
```
