```python
import os
import pandas as pd
import numpy as np
import re
```

```python
# load datasets 
df = pd.read_csv('Dataset_2.csv')
df
```

```python
# clean up dataframe by formatting all to strings 
# df[['PhoneNumber1','PhoneNumber2']] = df[['PhoneNumber1','PhoneNumber2']].astype(str)
df['PhoneNumber1'] = df['PhoneNumber1'].apply(lambda x: re.sub("[^0-9]", "", str(x)))
df['PhoneNumber2'] = df['PhoneNumber2'].apply(lambda x: re.sub("[^0-9]", "", str(x)))
```

```python
unique_customers = df['CustomerID'].unique()
number_list = []

for k in unique_customers:
    selected_data = df[df['CustomerID'] == k].reset_index(drop=True)
    if (len(selected_data['PhoneNumber1'][0]) == 10) & (selected_data['PhoneNumber1'][0].isnumeric()):
        number_list.append(selected_data['PhoneNumber1'][0])
    elif (len(selected_data['PhoneNumber2'][0]) == 10) & (selected_data['PhoneNumber2'][0].isnumeric()):
        number_list.append(selected_data['PhoneNumber2'][0])
    else:
        number_list.append(np.nan)
```

```python
output_df = pd.DataFrame()
output_df['CustomerID'] = unique_customers
output_df['PhoneNumber'] = number_list
output_df
```
